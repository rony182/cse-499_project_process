```python
# Import lybraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
import sqlite3

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('investopedia.db')

# Create a cursor object
cursor = conn.cursor()

# Create table to store the investopedia
cursor.execute('''
CREATE TABLE IF NOT EXISTS investopedia (
    id INTEGER PRIMARY KEY,
    url TEXT,
    article_pre_content TEXT,
    article_content TEXT,
    article_post_content TEXT
)
''')

# Commit changes and close the connection for now
conn.commit()
conn.close()
```

```python
def insert_article(url, pre_content, content, post_content):
    conn = sqlite3.connect('investopedia.db')
    cursor = conn.cursor()
    cursor.execute('''
    INSERT INTO investopedia (url, article_pre_content, article_content, article_post_content)
    VALUES (?, ?, ?, ?)
    ''', (url, pre_content, content, post_content))
    conn.commit()
    conn.close()
```

```python
# Define the URL of the page to scrape
url = 'https://www.investopedia.com/financial-term-dictionary-4769738'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
    'Host': 'www.investopedia.com',
    'Referer': 'https://www.google.com/',
    'Upgrade-Insecure-Requests': '1',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'none',
    'Sec-Fetch-User': '?1',
}
# Send a GET request to the URL
response = requests.get(url, headers=headers)
print(response.status_code)
print(response.text)
```

```python

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    # Find the main tag that contains the data
    main_tag = soup.find('main')
    # find the class that contains the data dictionary-top24-list__heading-link
    a_list = main_tag.find_all('a', class_='dictionary-top24-list__heading-link')
    #extract the href of the links
    href_list = [a['href'] for a in a_list]
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

```

```python
for link in href_list:
    response = requests.get(link, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    main_tag = soup.find('main')
    a_list = main_tag.find_all('a', class_='dictionary-top300-list__list')
    href_list_2 = [a['href'] for a in a_list]
    for link_2 in href_list_2:
        response = requests.get(link_2, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        main_tag = soup.find('main')
        #find the article tag that contains the data
        article_tag = main_tag.find('article')

        if article_tag:
            # Within the article tag, find the specified classes
            article_pre_content = article_tag.find('div', class_='article-pre-content')
            article_content = article_tag.find('div', class_='article-content')
            article_post_content = article_tag.find('div', class_='article-post-content')

            # Extract text content
            pre_content_html = str(article_pre_content) if article_pre_content else ''
            content_html = str(article_content) if article_content else ''
            post_content_html = str(article_post_content) if article_post_content else ''

            # Store the data in the database
            insert_article(link_2, pre_content_html, content_html, post_content_html)

        # Pause for 5 to 7 seconds
        time.sleep(3 + 2 * random.random())
        
    time.sleep(3 + 2 * random.random())
```
